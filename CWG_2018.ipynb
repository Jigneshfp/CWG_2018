```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import csv
f= csv.writer(open("CWG.csv", 'w'))
f.writerow(['pos', 'country', 'gold', 'silver', 'bronze', 'total'])
url = 'https://sports.ndtv.com/commonwealth-games-2018/medals-tally'

r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')

cwg_table = soup.find('tbody')
rows = cwg_table.find_all('tr')
```

```python
CWG_dic = {}
POS = []
COUNTRY = []
GOLD = []
SILVER = []
BRONZE = []
TOTAL = []
    
for item in rows:
    cells = item.find_all('td')
    pos = cells[0].text.strip()
    POS.append(pos)
    country = cells[1].text.strip()
    COUNTRY.append(country)
    gold = cells[2].text.strip()
    GOLD.append(gold)
    silver = cells[3].text.strip()
    SILVER.append(silver)
    bronze = cells[4].text.strip()
    BRONZE.append(bronze)
    total = cells[5].text.strip()
    TOTAL.append(total)
CWG_dic = {"POS":POS, "Country":COUNTRY, "Gold":GOLD, "Silver":SILVER, "Bronze":BRONZE, "Total":TOTAL}

    
```

```python
df = pd.DataFrame(CWG_dic, columns = CWG_dic.keys())
```

```python
df['Total'] = pd.to_numeric(df['Total'])
```

```python
from plotnine import *
```

```python
index = ['Gold', 'Silver', 'Bronze', 'Total']
for col in index:
    df[col] = pd.to_numeric(df[col])
df['Country'] = df['Country'].astype('category')
```

```python
#melting 3 columns gold, silver, bronze into single column : 'Medal'

df_new = pd.melt(df, id_vars=['POS', 'Country'], value_vars = ['Gold', 'Silver', 'Bronze'], var_name = 'Medal', value_name='No_ofMedal')
```

```python
df_new['Medal'] = df_new['Medal'].astype('category')
df_new['Country']= df_new['Country'].astype('category')
```
